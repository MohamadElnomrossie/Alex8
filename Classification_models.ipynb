```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
import matplotlib.pyplot as plt
#Step 2: Load the Dataset
#Scikit-learn conveniently includes the Iris dataset, so we can load it like this:
from sklearn.datasets import load_iris
iris = load_iris()
#The iris object is a dictionary-like object that contains the features, labels (target variable), and metadata. We can load these into a Pandas DataFrame for convenience:

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df
df['target'] = iris.target
#Step 3: Split the Data into Training and Test Sets
#We can use scikit-learn’s train_test_split function for this:

X_train, X_test, y_train, y_test = train_test_split(df[iris.feature_names], df['target'], random_state=0)
#Here, we’re using 75% of the data for training and 25% for testing by default.

#Step 4: Build the Decision Tree Model
#Now, we’re ready to create our decision tree. We’ll use scikit-learn’s DecisionTreeClassifier class:

clf = DecisionTreeClassifier(random_state=0)
clf.fit(X_train, y_train)
#Step 5: Make Predictions and Evaluate the Model
#We can now use our trained model to make predictions on our test data:

y_pred = clf.predict(X_test)
#And evaluate the accuracy of our model:

print("Accuracy:", accuracy_score(y_test, y_pred))
#Step 6: Visualize the Decision Tree
#Finally, let’s visualize the decision tree using scikit-learn’s plot_tree function:

fig = plt.figure(figsize=(15,10))
_ = tree.plot_tree(clf,
                   feature_names=iris.feature_names,
                   class_names=iris.target_names,
                   filled=True)
```
