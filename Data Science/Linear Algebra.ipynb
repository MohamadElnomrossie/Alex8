```python
def matrix_add(m1,m2):
    l1=[]
    
    for i in range(len(m1)):
        l2=[]
        for j in range(len(m1[i])):
            l2.append(m1[i][j]+m2[i][j])
        l1.append(l2)
    return l1
```

```python
matrix_add(m1,m2)
```
