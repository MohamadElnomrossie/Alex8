```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
X,y=fetch_california_housing(return_X_y=True)
```

```python
data=fetch_california_housing(as_frame=True)
```

```python
df=data['data']
y=data['target']*1000
X=df.values
scaler=MinMaxScaler()
X=scaler.fit_transform(X)
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

# Plain Linear Regression

```python
linear_clf=LinearRegression()
linear_clf.fit(X_train,y_train)
```

## Performance

```python
linear_clf.score(X_test,y_test)
```

```python
linear_clf.coef_
```

# Ridge

```python

ridge=Ridge(alpha=0.51)
ridge.fit(X_train,y_train)
ridge.score(X_test,y_test)
```

```python
alpha_values=np.linspace(0,1,50)
for i in alpha_values:
    ridge=Ridge(alpha=i)
    ridge.fit(X_train,y_train)
    print(ridge.score(X_test,y_test))
    print("i",i)
    print("="*80)
```

```python
ridge.coef_
```

# Lasso

```python
ls=Lasso()
ls.fit(X_train,y_train)
print(ls.score(X_test,y_test))
print(ls.coef_)
```

```python
alpha_values=np.linspace(0,1,50)
for i in alpha_values:
    ls=Lasso(alpha=i)
    ls.fit(X_train,y_train)
    print(ls.score(X_test,y_test))
    print("i",i)
    print("="*80)
```

# Elsatic Net

```python
el=ElasticNet()
el.fit(X_train,y_train)
print(el.score(X_test,y_test))
print(el.coef_)
```

```python
alpha_values=np.linspace(0,1,num=5)
ratios=np.linspace(0,1,num=5)
for i in alpha_values:
    for ratio in ratios:
        el=ElasticNet(alpha=i,l1_ratio=ratio)
        el.fit(X_train,y_train)
        print(el.score(X_test,y_test))
        print("i",i)
        print("="*80)
    print("*"*80)
```
